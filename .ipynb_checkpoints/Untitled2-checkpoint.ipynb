```python
import geopandas as gpd
```

```python
from shapely.geometry import box
import cartopy.crs as ccrs
import regionmask
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
```

```python
n=10

d = {
     "coords": {"lat": {"dims": ("var"), "data": np.arange(n)*np.random.rand(n)},
                "lon": {"dims": ("var"), "data": np.arange(n)*np.random.rand(n)+150}},
     "data_vars": {
         "var": {"dims": ("lon", "lat"), "data": np.array([np.random.rand(n) for i in range(10)])}},}
ds = xr.Dataset.from_dict(d)
ds
```
