# Effect of contour and contourf on plotting time

When there's lots of details in the plots, they will take lots of time to calculate, for example SPI change data here. That is because it's quite expensive to calculate the smoothing contours and the grey outlines.

Speed may be increased by :
1. reducing the intervals eg fewer ticks
2. turning off smoothing eg ```contourf = False```
3. turning off the grey contour lines, eg ```contour = False```

Choose what is relevant for you and your data.

````python
from acs_plotting_maps import plot_acs_hazard, regions_dict, cmap_dict, tick_dict
import xarray as xr
import numpy as np
````

````python
%%time
plot_file = '/g/data/ia39/ncra/drought_aridity/spi/downscaled_notBC_5km/SPI3_pct_time_below_-1_MME_change_GWL2.0_to_GWL1.2_percentiles_10-50-90.nc'
percentile = 0.5
data = xr.open_dataset(plot_file).sel(quantile=percentile).SPI3
````

````python
%%time
plot_acs_hazard(data =data,
                regions = regions_dict['ncra_regions'],
                cmap = cmap_dict["anom_b2r"],
                ticks = np.arange(-15, 15.1, 3),
                cbar_label = "Change in percent time\nSPI3  -1",
                cbar_extend = "both",
                title = f"Change in percent time of SPI3  -1 for MME {str(int(percentile*100))}th percentile",
                dataset_name = "Downscaled, not bias-corrected BARPA-R/CCAM runs from /g/data/ia39. ",
                date_range = f"From GWL1.2 to {plot_file[-40:-34]}",
               );
````
